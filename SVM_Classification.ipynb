<h2>Implementation of Support Vector Classification algorithm to classify the cases of breast cancer</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
dataset = pd.read_csv('Documents/AI/Lab_11/Datasets/breast_cancer_data.csv')
print(dataset.head())
```

```python
def diagnosis(x):
    if x=='M' :
        return 1
    if x=='B' :
        return 0
dataset['diagnosis'] = dataset['diagnosis'].apply(diagnosis)
print(dataset) 
```

```python
print("Any missing sample in data set:", dataset.isnull().values.any(), "\n") 
```

```python
dataset = dataset.replace([np.inf, -np.inf], np.nan)
dataset= dataset.fillna(dataset.mean())
dataset
```

```python
dataset= dataset.drop(columns=["Unnamed: 32"]) 
```

```python
Y = dataset['diagnosis']
X = dataset.drop(columns=['diagnosis'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2, random_state=9)
print('X train shape: ', X_train.shape)
print('Y train shape: ', Y_train.shape)
print('X test shape: ', X_test.shape)
print('Y test shape: ', Y_test.shape) 
```

```python
svc_classifier= SVC(kernel='poly') 
```

```python
svc_classifier.fit(X_train,Y_train)
```

```python
y_pred=svc_classifier.predict(X_test)
```

```python
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred)) 
```
